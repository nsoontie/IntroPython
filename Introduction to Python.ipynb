## 2. Functions
* Use functions to save blocks of code for later use.
* Use functions to make programs easier to understand.

### Defining functions
* A function looks like `def function_name(parameters):` followed by an indented block of code. 
* The function's *parameters* are listed in the parentheses and separated by commas if there are multiple parameters.
* The code executed by a function is indented.
* Functions with no parameters have empty parentheses.
* Functions can return a result with a return statement.
 

`````python
def fahr_to_kelvin(temp):
    temp_kelvin = ((temp - 32)*5/9) + 273.15
    return temp_kelvin
`````

### Calling functions
* Call a function using the function's name and passing parameters in the parantheses. 
* Parameters are assigned to temporary variables during the function's execution.

`````python
temperature_kelvin = fahr_to_kelvin(32)
`````

`````python
print(temperature_kelvin)
`````

`````python
def hello():
    print('Hello')
`````

`````python
hello()
`````

### Documenting functions
* Document functions so that you remember how to use them. 
* Even the simplest functions should be documented.
* Define a 'Docstring' in functions with three quotes.
* A function's 'Docstring' can be viewed with the `help` command.

`````python
def fahr_to_kelvin(temp):
    """Convert temperature in Fahrenheit to Kelvin.
    inputs: temp - the temperature in Fahrenheit
    returns: the temperature in Kelvin
    """
    temp_kelvin = ((temp - 32)*5/9) + 273.15
    return temp_kelvin
`````

`````python
help(fahr_to_kelvin)
`````

### Composing functions
* Functions can be called inside other functions

`````python
def kelvin_to_celsius(temp):
    """Convert temperature in Kelvin to Celsius.
    inputs: temp - the temperature in Kelvin
    returns: the temperature in Celsius
    """
    return temp - 273.15
`````

`````python
print(kelvin_to_celsius(273.15))
`````

### Exercise

Using the `kelvin_to_celsius` and `fahr_to_kelvin` functions defined above, write a function to convert a temperature from Fahrenheit to Celsius.

#### Solution

`````python
def fahr_to_celsius(temp):
    """Convert the temperture in Fahrenheit to Celsius.
    inputs: temp - the temperature in Fahrenheit
    returns: the temperature in Celsius
    """
    temp_kelvin = fahr_to_kelvin(temp)
    return kelvin_to_celsius(temp_kelvin)
`````

`````python
print(fahr_to_celsius(32))
`````

# 3. Importing libraries
* A library is a collection of files (called modules) that contain functions and constants for use in your program.
* Sometimes the term module and library are used interchangably, particularly for libraries with only one module.
* A library must be imported before it can be used in your program.
* Use `import` to load a library module.
* Refer to things from the module using `module_name.thing_name`.
* `math` is a library for mathematical functions

`````python
import math

print(math.cos(0))
print(math.cos(math.pi))
`````

* Modules can be imported with an alias or shorter name.
    * Refer to things from the module with `alias_name.thing_name`
* Use `help()` to display the docstrings from function in a module.

### Loading data
* Use `numpy.loadtxt()` to load data from a text file.
* The `skiprows` parameter allows you to skip `n` rows starting from the top of the text file.
* Use a variable to save the output of `loadtxt` for later use.
* The output of `numpy.loadtxt()` is an object called `numpy.ndarray` (also called a numpy array).

`````python
array = np.loadtxt('RaceRocksDailySalTemp.txt', skiprows=3 )
`````

`````python
type(array)
`````

* Numpy array's have attributes  (like `shape`, `size`, `ndim`). You can think of these as properties of the array.

`````python
print(array.shape)
print(array.size)
print(array.ndim)
`````

* You can access elements of an array with indexing. Indices start from 0.
* Indices are written in square brackets and indices for each axis or dimension are separated by columns.
* Use slicing with the `:` to access mulitple elements at once.

`````python
print('Element in row 0, column 0:', array[0,0])
print('Element in row 0, column 4:', array[1,4])
print('Elements in row 0:', array[0,:])
`````

* Numpy arrays also have methods (like `array.max()`, `array.min()`, `array.mean()`). You can think of these as array actions or functions.

`````python
print(array.max())
print(array.min())
print(array.mean(axis=0))
`````

### Masking data
* Use masked numpy arrays can mask invalid data
* Masked arrays do not used masked values in arithmetic operations.

`````python
array_masked = np.ma.masked_values(array, 99.9)
print(array_masked[0,:])
`````

`````python
print(array_masked.mean(axis=0))
`````

## 4. Analyzing time series data

Although `numpy` is a powerful tool for manipulating numerical data, there are even better tools for analyzing tabular data like that in `RaceRocksDailySalTemp.txt`. One example of a powerful tool is `pandas`.

### Pandas
* Pandas is a library for working with tabular data and other labelled data structures.

`````python
import pandas as pd
`````

#### Loading data with pandas
* A `DataFrame` is a `pandas` data structure for organizing tabular data. 
* A `DataFrame` has rows (labelled by an index) and columns (lablled by a header).
* Load tabular data from a text file with a pandas function `pandas.read_table()`.
* `pandas.read_table()` has many keyword arguments. We will use:
  * `header=2` indicates that row 2 contains column names (remember counting starts at 0)
  * `delim_whitespace=True` siginifies that data values are separated by white space
  * `parse_dates = {'Date': [0,1,2]}` attempts to parse information in columns 0,1, and 2 as dates. A new column is created with heading 'Date'.
  * `keep_date_col=True` specififies to keep the original columns if `parse_dates` was used.
  * `na_value=99.9` indicates to treat `99.9` as `NaN`.

`````python
help(pd.read_table)
`````

`````python
data = pd.read_table('RaceRocksDailySalTemp.txt', header=2, delim_whitespace=True, 
                     parse_dates={'Date': [0,1,2]}, keep_date_col=True, na_values=99.9)
`````

`````python
type(data)
`````

`````python
print(data)
`````

#### Organizing data
* pandas can reorganize data using 'Split-Apply-Combine' techniques.
* Use `DataFrame.dropna()` to remove nans from dataframe.
    * The `subset` keyword allows you to list which columns to consider.
* Determine which data points have temperature above the mean temperature.
* Determine percentage of data points when temperature was above mean.
* Apply `DataFrame.describe()` to new data frames.

`````python
data_nonan_temp = data.dropna(subset=['Temperature(C)'])
mean_temp = data_nonan_temp['Temperature(C)'].mean()
mask_higher = data_nonan_temp['Temperature(C)'].apply(lambda x:x>mean_temp)
mask_higher.aggregate('sum', axis=0)/len(mask_higher)
`````

`````python
mask_higher.describe()
`````

* Use `DataFrame.groupby()` to organize data by common value.
* Plot time series of annual mean temperture.

`````python
data_year=data_nonan_temp.groupby(['Year'])
data_year['Temperature(C)'].mean().plot()
`````

* Plot a monthly climatology
* Convert `DataFrame` index from string to integer
* Sort `DataFrame` index with `sort_index()`.
* Reproduce temperature-salinty plot with month color coded.

`````python
data_month = data_nonan_temp.groupby(['Month'], sort=True)
data_month.describe()
`````

`````python
monthly_clim = data_month['Temperature(C)'].mean()
monthly_clim.index = monthly_clim.index.map(int)
monthly_clim = monthly_clim.sort_index()
monthly_clim.plot()
`````

`````python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
fig,ax = plt.subplots(1,1)
colors=cm.plasma(np.linspace(0,1,12))
months=range(1,13)
for month in months:
    df_month=data_month.get_group(str(month))
    df_month.plot(x='Salinity(psu)', y='Temperature(C)', kind='scatter', 
                  label=month,color=colors[month-1], ax=ax,alpha=0.5)
`````
